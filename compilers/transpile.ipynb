```python
from qiskit import QuantumCircuit
from qiskit.compiler import transpile
```

```python
circuit = QuantumCircuit(3, 3)
circuit.ccx(0,1,2)
circuit.measure([0,1, 2], [0,1, 2])
circuit.draw('mpl')
```

```python
circuit_ibm = transpile(circuit, basis_gates = ['cz', 'sx', 'rz'], coupling_map =[[0, 1], [1, 2]] , optimization_level=1)
circuit_ibm.draw('mpl')
```

```python
circuit_ibm = transpile(circuit, basis_gates = ['cz', 'sx', 'rz'], coupling_map =[[0, 1], [1, 2]] , optimization_level=2)
circuit_ibm.draw('mpl')
```

```python
circuit_ibm = transpile(circuit, basis_gates = ['cz', 'sx', 'rz'], coupling_map =[[0, 1], [1, 2]] , optimization_level=3)
circuit_ibm.draw('mpl')
```
