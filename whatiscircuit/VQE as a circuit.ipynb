```python
from qiskit.circuit import QuantumCircuit, ParameterVector, ClassicalRegister

num_qubits = 2
phi = ParameterVector('phi', length=num_qubits)
theta = ParameterVector('theta', length=num_qubits)


circuit = QuantumCircuit(num_qubits)
circuit.reset(range(num_qubits))
circuit.barrier(range(num_qubits))
[circuit.rx(p,i) for i, p in enumerate(theta)]
[circuit.rz(p,i) for i, p in enumerate(phi)]
circuit.cx(0,1)
circuit.barrier(range(num_qubits))
circuit.draw(output='mpl')
```

```python
#xz
circuit = QuantumCircuit(num_qubits, num_qubits)
circuit.reset(range(num_qubits))
circuit.barrier(range(num_qubits))
[circuit.rx(p,i) for i, p in enumerate(theta)]
[circuit.rz(p,i) for i, p in enumerate(phi)]
circuit.cx(0,1)
circuit.barrier(range(num_qubits))
circuit.h(1)
circuit.measure(range(num_qubits),range(num_qubits))
circuit.draw(output='mpl')
```

```python
#zy
circuit = QuantumCircuit(num_qubits, num_qubits)
circuit.reset(range(num_qubits))
circuit.barrier(range(num_qubits))
[circuit.rx(p,i) for i, p in enumerate(theta)]
[circuit.rz(p,i) for i, p in enumerate(phi)]
circuit.cx(0,1)
circuit.barrier(range(num_qubits))
circuit.s([0])
circuit.h([0])
circuit.measure(range(num_qubits),range(num_qubits))
circuit.draw(output='mpl')
```

```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
cregx = ClassicalRegister(4, 'step_{1}')
cregz = ClassicalRegister(4, 'step_{2}')
creg3 = ClassicalRegister(4, 'step_{3}')
qreg = QuantumRegister(2, 'q')
circuit = QuantumCircuit(qreg,cregx, cregz, creg3)
circuit.reset(range(num_qubits))
circuit.barrier(range(num_qubits))
[circuit.rx(p,i) for i, p in enumerate(theta)]
[circuit.rz(p,i) for i, p in enumerate(theta)]
circuit.cx(0,1)
circuit.barrier(range(num_qubits))
circuit.h(1)
circuit.measure(range(num_qubits),[cregz[0],cregz[1]])
circuit.barrier(range(num_qubits))
circuit.reset(range(num_qubits))
circuit.barrier(range(num_qubits))
[circuit.rx(p,i) for i, p in enumerate(theta)]
[circuit.rz(p,i) for i, p in enumerate(theta)]
circuit.cx(0,1)
circuit.barrier(range(num_qubits))
circuit.s([0])
circuit.h([0])
circuit.measure(range(num_qubits),[cregz[2],cregz[3]])
circuit.barrier(range(num_qubits))
circuit.reset(range(num_qubits))
circuit.barrier(range(num_qubits))
[circuit.rx(p,i) for i, p in enumerate(theta)]
[circuit.rz(p,i) for i, p in enumerate(theta)]
circuit.cx(0,1)
circuit.barrier(range(num_qubits))
circuit.h(1)
circuit.measure(range(num_qubits),[creg3[0],creg3[1]])
circuit.barrier(range(num_qubits))
circuit.reset(range(num_qubits))
circuit.barrier(range(num_qubits))
[circuit.rx(p,i) for i, p in enumerate(theta)]
[circuit.rz(p,i) for i, p in enumerate(theta)]
circuit.cx(0,1)
circuit.barrier(range(num_qubits))
circuit.s([0])
circuit.h([0])
circuit.measure(range(num_qubits),[creg3[2],creg3[3]])

circuit.draw(output='mpl')
```
